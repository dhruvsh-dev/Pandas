```python
import numpy as np
import pandas as pd
```

## Finding Missing Data

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df=pd.DataFrame(data)
df
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

## Removing Missing Data

```python
df.dropna()
```

```python
df.dropna(thresh=3) # check non -null value
```

## Filling The Missing Data

```python
df.fillna(0)
```

```python
values={'A':0,'B':100,'C':200,'D':400}
df.fillna(value=values)
```
